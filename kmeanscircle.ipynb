```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Machine-Learning-Workshop/master/Chapter02/Exercise2.01/circles.csv'
data = pd.read_csv(file_url)

data.head(10)
plt.scatter(data.iloc[:,0], data.iloc[:,1])
```

```python

est_kmeans = KMeans(n_clusters=5, random_state=0)
est_kmeans.fit(data)
pred_kmeans = est_kmeans.predict(data)

plt.scatter(data.iloc[:,0], data.iloc[:,1], c=pred_kmeans)
plt.show()

    
```

```python
from sklearn.cluster import MeanShift

est_meanshift = MeanShift(bandwidth=0.5)
est_meanshift.fit(data)
est_pred = est_meanshift.predict(data)
plt.scatter(data.iloc[:,0], data.iloc[:,1], c=est_pred)
```

```python
from sklearn.cluster import DBSCAN
est_dbscan = DBSCAN(eps=0.1)
dbs_pred = est_dbscan.fit_predict(data)

plt.scatter(data.iloc[:,0], data.iloc[:,1], c=dbs_pred, s=20)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score

k_meansscore = silhouette_score(data, pred_kmeans, metric='euclidean')
meanshiftscore = silhouette_score(data, est_pred, metric='euclidean')
dbscanscore = silhouette_score(data, dbs_pred, metric='euclidean')

print(k_meansscore, meanshiftscore, dbscanscore)
```

```python
k_meansscore = calinski_harabasz_score(data, pred_kmeans)
meanshiftscore = calinski_harabasz_score(data, est_pred)
dbscanscore = calinski_harabasz_score(data, dbs_pred)
print(k_meansscore, meanshiftscore, dbscanscore)
```
